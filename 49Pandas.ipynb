```python
# # Self defined function
# 1) Calculate total marks for each student
# 2) Calculate  percentage for each student
# 3) Give Grade according to marks 
# Above 90% Grade "O"
# Above 80% Grade "A+"
# Above 70% Grade "A"
# Above 60% Grade "B+"
# Above 50% Grade "B"
# Above 40% Grade "C+"
# Above 35% Grade "C"
# Otherwise Grade "F"
```

```python
a.columns
```

```python
def cal_result(a):
    a['Total marks'] = a['English']+a['Physics']+ a['Chemistry']+a['Mathematics']
    a['Percentage'] = (a['Total marks']/400)*100
    if a['Percentage']>90:
        a['grade'] = 'O'
    elif a["Percentage"]>80:
       a["Grade"]="A+"
    elif a["Percentage"]>70:
        a["Grade"]="A"        
    elif a["Percentage"]>60:
        a["Grade"]="B+"        
    elif a["Percentage"]>50:
        a["Grade"]="B"         
    elif a["Percentage"]>40:
        a["Grade"]="C"         
    elif a["Percentage"]>35:
        a["Grade"]="C+"         
    else:
        a["Grade"]="F"   
        
    return a
```
